```python
import numpy as np
from scipy import stats

# Marks of two classes
class_A = np.array([70, 75, 80, 85, 90])
class_B = np.array([60, 65, 70, 75, 80])

t_stat, p_value = stats.ttest_ind(class_A, class_B)

print("t-statistic:", t_stat)
print("p-value:", p_value)


before = np.array([50, 55, 60, 65, 70])
after = np.array([55, 60, 65, 70, 75])

t_stat, p_value = stats.ttest_rel(before, after)

print("Paired t-test p-value:", p_value)

# Test if average marks = 75
data = np.array([70, 72, 78, 80, 76])

t_stat, p_value = stats.ttest_1samp(data, 75)

print("One-sample t-test p-value:", p_value)

# Observed preferences
observed = np.array([30, 20])  # Tea, Coffee

# Expected (equal preference)
expected = np.array([25, 25])

chi_stat, p_value = stats.chisquare(observed, expected)

print("Chi-square statistic:", chi_stat)
print("p-value:", p_value)

# Example: Gender vs Preference
# Rows: Male, Female | Columns: Tea, Coffee
data = np.array([[20, 10],
                 [15, 25]])

chi_stat, p_value, dof, expected = stats.chi2_contingency(data)

print("Chi-square statistic:", chi_stat)
print("p-value:", p_value)
print("Expected frequencies:\n", expected)


# Result Interpretation
alpha = 0.05

if p_value < alpha:
    print("Result is statistically significant")
else:
    print("Result is NOT statistically significant")
```
