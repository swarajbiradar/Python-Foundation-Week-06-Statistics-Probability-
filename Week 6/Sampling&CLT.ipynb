```python
import numpy as np
import matplotlib.pyplot as plt

# Population of heights (mean=170, std=10)
population = np.random.normal(loc=170, scale=10, size=10000)

plt.hist(population, bins=30)
plt.title("Population Distribution (Heights)")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()


# Task 1: Draw a Random Sample
sample = np.random.choice(population, size=50)

print("Sample Data:", sample[:10])
print("Sample Mean:", np.mean(sample))


# Task 2: Take Multiple Samples and Compute Means
sample_means = []

for i in range(100):
    sample = np.random.choice(population, size=50)
    sample_means.append(np.mean(sample))

print("First 10 Sample Means:", sample_means[:10])


#   Task 3: Plot Distribution of Sample Means (CLT)
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()


#Task 4: Compare Different Sample Sizes
means_small = []
means_large = []

# Small sample size
for i in range(100):
    sample = np.random.choice(population, size=10)
    means_small.append(np.mean(sample))

# Large sample size
for i in range(100):
    sample = np.random.choice(population, size=100)
    means_large.append(np.mean(sample))

plt.hist(means_small, bins=30, alpha=0.5)
plt.hist(means_large, bins=30, alpha=0.5)

plt.title("Effect of Sample Size on CLT")
plt.show()

# Task 5: Compare Population Mean vs Sample Mean
population_mean = np.mean(population)
sample_mean = np.mean(sample_means)

print("Population Mean:", population_mean)
print("Average of Sample Means:", sample_mean)
```
