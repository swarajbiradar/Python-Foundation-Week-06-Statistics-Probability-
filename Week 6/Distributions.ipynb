```python
import numpy as np
import matplotlib.pyplot as plt
from setuptools import Distribution

# Generate exam scores (mean=70, std=10)
scores = np.random.normal(loc=70, scale=10, size=1000)

plt.hist(scores, bins=30)
plt.title("Normal Distribution of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()

# Generate another normal distribution
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30)
plt.title("Standard Normal Distribution")
plt.show()

# Simulate 10 coin tosses repeated 1000 times
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)

plt.hist(binomial_data, bins=10)
plt.title("Binomial Distribution (Coin Toss)")
plt.xlabel("Number of Heads")
plt.ylabel("Frequency")
plt.show()

# Average 5 calls per hour
poisson_data = np.random.poisson(lam=5, size=1000)

plt.hist(poisson_data, bins=15)
plt.title("Poisson Distribution (Calls per Hour)")
plt.xlabel("Number of Calls")
plt.ylabel("Frequency")
plt.show()

normal = np.random.normal(50, 5, 1000)
binomial = np.random.binomial(20, 0.5, 1000)
poisson = np.random.poisson(5, 1000)

plt.hist(normal, bins=30, alpha=0.5)
plt.hist(binomial, bins=20, alpha=0.5)
plt.hist(poisson, bins=20, alpha=0.5)

plt.title("Comparison of Distributions")
plt.show()
```
