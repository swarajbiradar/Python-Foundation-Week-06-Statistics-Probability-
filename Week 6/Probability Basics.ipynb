```python
#Task 1: Probability of Rolling a 4 on a Dice


prob = 1 / 6

print("Probability of rolling a 4:", prob)

#Task 2: Probability of Getting Heads in a Coin Toss
prob = 1 / 2

print("Probability of getting Heads:", prob)

#Task 3: Probability of Getting Heads Twice

prob = (1/2) * (1/2)

print("Probability of getting Heads twice:", prob)

#Task 4: Probability of Getting Even Number on Dice

prob = 3 / 6

print("Probability of even number:", prob)

#Task 5: Addition Rule (Mutually Exclusive Events)

prob = (1/6) + (1/6)

print("Probability of getting 1 or 2:", prob)
```

```python
import numpy as np

# Simulate 1000 dice rolls
rolls = np.random.randint(1, 7, 1000)

# Count how many times 4 appears
count_4 = np.sum(rolls == 4)

prob = count_4 / 1000

print("Experimental Probability of 4:", prob)
```
