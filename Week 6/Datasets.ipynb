```python
import numpy as np

#Dataset (Exam Scores)
scores = np.array([70, 75, 80, 85, 90, 95, 60, 65, 88, 92])

print("Scores:", scores)

#Task 1: Calculate Mean
mean_value = np.mean(scores)

print("Mean (Average):", mean_value)

#Task 2: Calculate Median
median_value = np.median(scores)

print("Median:", median_value)

#Task 3: Calculate Variance
variance_value = np.var(scores)

print("Variance:", variance_value)

#Task 4: Calculate Standard Deviation
std_dev = np.std(scores)

print("Standard Deviation:", std_dev)

#Task 5: Compare Two Datasets
dataset1 = np.array([50, 60, 70, 80, 90])
dataset2 = np.array([70, 72, 74, 76, 78])

print("Dataset 1 Std Dev:", np.std(dataset1))
print("Dataset 2 Std Dev:", np.std(dataset2))


#Step-by-Step Variance Calculation (Manual)
mean = np.mean(scores)

# Step 1: subtract mean
deviations = scores - mean

# Step 2: square deviations
squared = deviations ** 2

# Step 3: average of squared values
variance_manual = np.mean(squared)

print("Manual Variance:", variance_manual)
```
